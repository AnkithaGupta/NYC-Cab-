```python
merged_df=pd.read_csv('/Users/Daniellaviale/Downloads/Dataviz_data_v2.csv', index_col=False)
```

```python
logistic_df = merged_df
```

```python
logistic_df.head()
```

```python
#Create dummies for the 'PUBorough' column
borough_dummies = pd.get_dummies(logistic_df['PUBorough'])
logistic_df = pd.concat([logistic_df, borough_dummies], axis=1)
logistic_df.drop('PUBorough', axis=1, inplace=True)
```

```python
#logistic_df.isnull().sum()
```

```python
logistic_df=logistic_df.drop(['Unnamed: 0'],axis=1)
logistic_df=logistic_df.drop(['month'],axis=1)
```

```python
#logistic_df['day'].value_counts()
```

```python
day_map = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday',
           5: 'Friday', 6: 'Saturday', 0: 'Sunday'}

# replace values in the day column using the dictionary
logistic_df['day'] = logistic_df['day'].replace(day_map)
```

```python
logistic_df['day'].value_counts()
```

```python
#Create dummies for the 'Day' column
day_dummies = pd.get_dummies(logistic_df['day'])
logistic_df = pd.concat([logistic_df, day_dummies], axis=1)
logistic_df.drop('day', axis=1, inplace=True)
```

```python
logistic_df['hour'].value_counts()
```

```python
logistic_df.head()
```

```python
hour_map = {0: 'Midnight', 1: 'Midnight', 2: 'Midnight', 3: 'Midnight', 4: 'Morning', 5: 'Morning', 6: 'Morning', 7: 'Morning',
           8: 'Mid_morning', 9: 'Mid_morning', 10: 'Mid_morning', 11: 'Mid_morning', 12: 'Midday', 13: 'Midday', 14: 'Midday', 15: 'Midday',
        16: 'Evening', 17: 'Evening', 18: 'Evening', 19: 'Evening', 20: 'Night', 21: 'Night', 22: 'Night', 23: 'Night'}

# replace values in the day column using the dictionary
logistic_df['hour'] = logistic_df['hour'].replace(hour_map)
logistic_df.drop('hour', axis=1, inplace=True)

#Midnight: 0, 1, 2, 3
#Morning: 4, 5, 6, 7
#Mid_morning: 8, 9, 10, 11
#Midday: 12, 13, 14, 15
#Evening: 16, 17, 18, 19
#Night: 20, 21, 22, 23
```

```python
#Create dummies for the 'Hour' column
hour_dummies = pd.get_dummies(logistic_df['hour'])
logistic_df = pd.concat([logistic_df, hour_dummies], axis=1)
#logistic_df.drop('hour', axis=1, inplace=True)
```

```python
logistic_df['hour'].value_counts()
```

```python
#Using Staten Island, Midnight, Sunday as base category
x=logistic_df[['trip_miles','base_passenger_fare','Morning','Mid_morning','Midday','Night','Evening','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Bronx','Brooklyn','EWR', 'Manhattan','Queens']]
y=logistic_df['highvolume']

##train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=101)

#Implementing the Log Model
from sklearn.linear_model import LogisticRegression #import
logmodel=LogisticRegression(solver='liblinear') #initialize
logmodel.fit(x_train,y_train)#fit

logmodel.intercept_ #bo
logmodel.coef_ #b1

probabilities=logmodel.predict_proba(x_test)
y_pred=logmodel.predict(x_test)

##evaluate the model using f1-score
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
from sklearn.metrics import (accuracy_score, precision_score,recall_score,f1_score,confusion_matrix)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
from scipy.stats import norm
import numpy as np
# create a DataFrame to store the feature names, coefficients, and p-values
coef_df = pd.DataFrame({'feature': x.columns, 'coefficient': logmodel.coef_[0]})

# calculate the p-values for each coefficient using the normal distribution
coef_df['p-value'] = [2 * (1 - norm.cdf(np.abs(x))) for x in logmodel.coef_[0]]

# print the coefficients, p-values, and significance levels (alpha=0.05) next to the feature names
for idx, row in coef_df.iterrows():
    if row['p-value'] < 0.05:
        signif = '*'
    else:
        signif = ''
    print(f"{row['feature']}: {row['coefficient']:.3f} (p-value={row['p-value']:.3f}{signif})")
```
