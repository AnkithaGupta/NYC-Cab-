### Reading the location. look up data

```python
data_loc = pd.read_csv("taxi+_zone_lookup.csv")
```

```python
data = pd.merge(data,data_loc , left_on = 'PULocationID', right_on = 'LocationID', how = 'inner')
```

```python
data = data.rename(columns= {'Borough':'PUBorough','Zone':'PUZone','service_zone':'PUservice_zone'})
```

```python
data.head()
```

data.dtypes

```python
data['pickup_datetime'] =  pd.to_datetime(data['pickup_datetime'])
```

```python
data['dropoff_datetime'] =  pd.to_datetime(data['dropoff_datetime'])
```

```python
data['hour'] =data['pickup_datetime'].dt.hour
```

```python
data.head()
```

```python
data.hour.unique()
```

```python
data['day'] = data['pickup_datetime'].dt.weekday
```

```python
data['month'] = data['pickup_datetime'].dt.month
```
